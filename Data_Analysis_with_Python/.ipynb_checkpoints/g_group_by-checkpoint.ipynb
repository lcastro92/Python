```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.DataFrame()
df["drive-wheels"] = None
df["body-style"]=None
df["price"]=None
drive=["4wd","4wd","4wd","fwd","fwd","fwd","fwd","fwd","rwd","rwd","rwd","rwd","rwd"]

body=["hatchback","sedan","wagon","convertible","hardtop","hatchback","sedan","wagon","convertible","hardtop","hatchback","sedan","wagon"]

price=[7603,12647,9095,11595,8249,8396,9611,9997,23949,24202,14337,21711,16994]

df["drive-wheels"]=drive
df["body-style"]=body
df["price"]=price


df
```

```python
#para agrupar hacemos
df_agrupado=df.groupby(['drive-wheels'],as_index=False).mean()
df_agrupado
```

```python
#Tambi√©n podemos hacer una tabla pivotante
df_pivot=df.pivot(index="drive-wheels",columns="body-style")
df_pivot
```

```python
#Otra  manera de representar una tabla pivotante, es con mapas de calor
plt.pcolor(df_pivot, cmap='RdBu')
plt.colorbar()
plt.show()
```
