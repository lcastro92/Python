```python
#Las rest apis permiten la comunicación a través de internet
#para utilizar recursos, algoritmos, etc
#REST = REpresentational State Transfer APIs

#La  api de la nba sirve para descargar data de la NBA
#(instalamos la api de nba con pip3 install nba_api)
import nba_api
import pandas as pd
```

```python
from nba_api.stats.static import teams
nba_teams=teams.get_teams()
nba_teams[:5]
```

```python
#Podemos transformar ese diccionario en una tabla
def one_dict(list_dict):
    keys=list_dict[0].keys()
    out_dict={key:[] for key in keys}
    for dict_ in list_dict:
        for key,value in dict_.items():
            out_dict[key].append(value)
    return out_dict

dict_nba_team=one_dict(nba_teams)
```

```python
#y esa tabla transformarla a un dataframe
dataframe=pd.DataFrame(dict_nba_team)
dataframe.head()
```

```python
#podemos hacer búsquedas, por ejemplo buscar los que tengan nickname warriors
df_warriors=dataframe[dataframe['nickname']=='Warriors']
df_warriors
```

```python
#guardamos el id en otra variable
id_warriors=df_warriors[['id']].values[0][0]
print(id_warriors)
```

```python
#Para poder buscar ese id en la base de datos, usamos la función leagueGameFinder
from nba_api.stats.endpoints import leaguegamefinder

gamefinder=leaguegamefinder.LeagueGameFinder(team_id_nullable=id_warriors)
```

```python
#ahora que tenemos la respuesta de la Api, en forma de HTML+JSON, usamos 
#la función de gamefinder para transformar a dataframe
games=gamefinder.get_data_frames()[0]
games.head()
```
