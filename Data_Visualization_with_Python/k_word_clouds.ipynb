```python
import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
```

```python
#El paquete para crear word clouds es word_cloud
# install wordcloud
!pip3 install wordcloud==1.8.1

# import package and its set of stopwords
from wordcloud import WordCloud, STOPWORDS

print ('Wordcloud is installed and imported!')
```

```python
#Vamos a analizar alicia en el país de las Maravillas
import urllib

#Abrimos el archivo y lo guardamos en una variable
alice=urllib.request.urlopen('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/alice_novel.txt').read().decode("utf-8")
```

```python
#Seleccionamos las palabras redundantes
stopwords = set(STOPWORDS)
```

```python
#Creamos el objeto wordcloud
alice_wc = WordCloud(
    background_color='white',
    stopwords=stopwords
)

#lo inicializamos
alice_wc.generate(alice)
```

```python
#Lo mostramos
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#Modificamos los tamaños
fig = plt.figure(figsize=(14, 18))
# mostramos la nube
plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#Agregamos "said" a la lista de palabras baneadas
stopwords.add('said') 

# regeneramos la nube
alice_wc.generate(alice)

# mostramos la nube
fig = plt.figure(figsize=(14, 18))

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
#Podemos agregar máscaras a las nubes
alice_mask = np.array(Image.open(urllib.request.urlopen('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/labs/Module%204/images/alice_mask.png')))
```

```python
#para ver la máscara
fig = plt.figure(figsize=(14, 18))

plt.imshow(alice_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
# Definimos la nube
alice_wc = WordCloud(background_color='white', max_words=2000, mask=alice_mask, stopwords=stopwords)

# La generamos
alice_wc.generate(alice)

# La mostramos
fig = plt.figure(figsize=(14, 18))

plt.imshow(alice_wc, interpolation='bilinear')
plt.axis('off')
plt.show()
```
