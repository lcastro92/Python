```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
filename='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx' 
df_can=pd.read_excel( filename, sheet_name="Canada by Citizenship", skiprows=range(20), skipfooter=2 )

df_can.rename(columns={"OdName":"Country"},inplace=True)
df_can.set_index("Country",inplace=True)
df_can["Total"]=df_can[1980]+df_can[1981]+df_can[1982]+df_can[1983]+df_can[1984]+df_can[1985]+df_can[1986]+df_can[1987]+df_can[1988]+df_can[1989]+df_can[1990]+df_can[1991]+df_can[1992]+df_can[1993]+df_can[1994]+df_can[1995]+df_can[1996]+df_can[1997]+df_can[1998]+df_can[1999]+df_can[2000]+df_can[2001]+df_can[2002]+df_can[2003]+df_can[2004]+df_can[2005]+df_can[2006]+df_can[2007]+df_can[2008]+df_can[2009]+df_can[2010]+df_can[2011]+df_can[2012]+df_can[2013]
df_can.head()
```

```python
# Tomamos la población total por año
years=list(range(1980,2014))
df_tot = pd.DataFrame(df_can[years].sum(axis=0))

# Cambiamos el tipo de dato a float
df_tot.index = map(float, df_tot.index)

# sacamos el total del índice
df_tot.reset_index(inplace=True)

# renombramos las columnas
df_tot.columns = ['year', 'total']

# mostramos los primeros 5 valores
df_tot.head()
```

```python
#Para generar el gráfico de regresion
sns.regplot(x='year', y='total', data=df_tot)
```

```python
#para cambiar los colores
sns.regplot(x='year', y='total', data=df_tot, color='green')
plt.show()
```

```python
#para cambiar la forma de los puntos
ax = sns.regplot(x='year', y='total', data=df_tot, color='green', marker='+')
plt.show()
```

```python
#para cambiar el tamaño
plt.figure(figsize=(15, 10))
sns.regplot(x='year', y='total', data=df_tot, color='green', marker='+')
plt.show()
```

```python
#Agregando títulos, etc
plt.figure(figsize=(15, 10))
ax = sns.regplot(x='year', y='total', data=df_tot, color='green', marker='+', scatter_kws={'s': 200})

ax.set(xlabel='Year', ylabel='Total Immigration') # add x- and y-labels
ax.set_title('Total Immigration to Canada from 1980 - 2013') # add title
plt.show()
```

```python
#cambiando la fuente
plt.figure(figsize=(15, 10))

sns.set(font_scale=1.5)

ax = sns.regplot(x='year', y='total', data=df_tot, color='green', marker='+', scatter_kws={'s': 200})
ax.set(xlabel='Year', ylabel='Total Immigration')
ax.set_title('Total Immigration to Canada from 1980 - 2013')
plt.show()
```
