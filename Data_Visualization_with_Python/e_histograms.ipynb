```python
import pandas as pd
```

```python
#los histogramas sirven para representar la distribución de la frecuencia de una variable
filename='https://s3-api.us-geo.objectstorage.softlayer.net/cf-courses-data/CognitiveClass/DV0101EN/labs/Data_Files/Canada.xlsx' 
df_can=pd.read_excel( filename, sheet_name="Canada by Citizenship", skiprows=range(20), skipfooter=2 )

df_can.head()
```

```python
df_can.rename(columns={"OdName":"Country"},inplace=True)
df_can.set_index("Country",inplace=True)
df_can["Total"]=df_can[1980]+df_can[1981]+df_can[1982]+df_can[1983]+df_can[1984]+df_can[1985]+df_can[1986]+df_can[1987]+df_can[1988]+df_can[1989]+df_can[1990]+df_can[1991]+df_can[1992]+df_can[1993]+df_can[1994]+df_can[1995]+df_can[1996]+df_can[1997]+df_can[1998]+df_can[1999]+df_can[2000]+df_can[2001]+df_can[2002]+df_can[2003]+df_can[2004]+df_can[2005]+df_can[2006]+df_can[2007]+df_can[2008]+df_can[2009]+df_can[2010]+df_can[2011]+df_can[2012]+df_can[2013]
df_can.head()
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
df_can[2013].plot(kind="hist")
plt.title("Histograma de inmigración de 195 países en 2013")
plt.ylabel("Paises")
plt.xlabel("Inmigrantes")
plt.show()
```

```python
#Como es dificil de interpretar la información, vamos a armar bins
count,bin_edges=np.histogram(df_can[2013])
df_can[2013].plot(kind="hist",xticks=bin_edges)
plt.title("Histograma de inmigración de 195 países en 2013")
plt.ylabel("Paises")
plt.xlabel("Inmigrantes")
plt.show()
```
