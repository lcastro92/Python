```python
import requests
import os

if not os.path.isfile("./Airport_Data.xlsx"): 
    r = requests.get("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/analysing-spreadsheet-data-with-python/Airport_Data.xlsx")
    f = open("./Airport_Data.xlsx", mode = "wb+")
    f.write(r.content)
    f.close()
```

Para leer archivos xsl usamos el método read de la librería de pandas

```python
import pandas as pd

df_facilities = pd.read_excel("./Airport_Data.xlsx", sheet_name="Facilities")
```

Podemos ver cómo se ve el archivo

```python
df_facilities
```

Podemos ver los tipos que hay en el archivo

```python
df_facilities["Type"]
```

Con unique podemos ver el listado sin repeticiones

```python
df_facilities["Type"].unique()
```

Podemos filtrar por valores específicos

```python
df_facilities["Type"] == "SEAPLANE BASE"
```

Podemos pasar como parámetro el valor para que se filtre en el dataset

```python
df_facilities[df_facilities["Type"] == "SEAPLANE BASE"]
```

Podemos ordenar por otros valores, como los aeropuertos más altos

```python
df_facilities.sort_values(by = "ARPElevation", ascending = False)
```

Para tomar 5 más altos podemos usar el método head

```python
df_facilities.sort_values(by = "ARPElevation", ascending = False).head(n = 5)
```

También podemos calcular el promedio de altura de los aeropuertos

```python
df_facilities["ARPElevation"].mean()
```

También calcular el máximo y el mínimo

```python
df_facilities["ARPElevation"].max()
```

```python
df_facilities["ARPElevation"].min()
```

Ahora podemos guardar nuestros resultados en un archivo xlsx

```python
df_seaplane = df_facilities[df_facilities["Type"] == "SEAPLANE BASE"]
df_seaplane.to_excel("./seaplane.xlsx")
```
